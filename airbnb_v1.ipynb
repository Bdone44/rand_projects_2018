### Model

```python
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics 
from sklearn.metrics import classification_report
```

```python
 contacts_all_v3.info()
```

#### What Impacts Booking?
All Users

```python
X =  contacts_all_v3[['contact_me_flag','new_user_flag','br_country_flag','location_unknown_flag','m_interactions','m_first_message_length_in_characters','m_guests','words_in_user_profile','room_apt_flag','room_private_flag','total_reviews_pos','interact_2_check_in_days','stay_duration_days']]
y = contacts_all_v3['book_flag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=25)
```

```python
total = X.isnull().sum().sort_values(ascending = False)
percent = (X.isnull().sum()/X.isnull().count()*100).sort_values(ascending = False)
missing_contacts  = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_contacts.head(20)
```

```python
LogReg = LogisticRegression()
LogReg.fit(X_train, y_train)
```

```python
y_pred = LogReg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
# examine the coefficients
pd.DataFrame(list(zip(X.columns, np.transpose(LogReg.coef_))))
```
